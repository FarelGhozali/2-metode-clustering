# Unsupervised Learning: K-Means vs Agglomerative Clustering
**Dataset:** Iris Species
**Tujuan:** Membandingkan efisiensi pengelompokan antara metode partisi (K-Means) dan hierarkis (Agglomerative).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# --- Data Acquisition ---
# Menggunakan dataset bawaan sklearn untuk menjamin interoperabilitas tanpa dependensi file eksternal.
data = load_iris()
X = pd.DataFrame(data.data, columns=data.feature_names)

print(f"[INFO] Data loaded. Shape: {X.shape}")
X.head()
```

```python
# --- Feature Engineering & Scaling ---
# Algoritma berbasis jarak (Euclidean) sangat sensitif terhadap skala fitur.
# StandardScaler memastikan semua fitur memiliki mean 0 dan variance 1.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reduksi dimensi menggunakan PCA menjadi 2 komponen utama.
# Tujuannya murni untuk visualisasi plot 2D agar hasil clustering bisa dipahami secara intuitif.
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

print("[INFO] Preprocessing & PCA dimension reduction complete.")
```

```python
# --- Model 1: K-Means (Centroid-Based) ---
# Kita set n_clusters=3 sesuai dengan struktur alami data Iris (Setosa, Versicolor, Virginica).
# n_init=10 digunakan untuk memastikan stabilitas centroid awal.
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X_scaled)

# Evaluasi menggunakan Silhouette Score: semakin mendekati 1, semakin solid klasternya.
km_score = silhouette_score(X_scaled, kmeans_labels)
print(f"Silhouette Score K-Means: {km_score:.4f}")

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=kmeans_labels, palette='viridis', s=60)
plt.title('K-Means Clustering Result')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(title='Cluster')

# --- Model 2: Agglomerative (Hierarchical-Based) ---
# Menggunakan pendekatan bottom-up tanpa inisialisasi centroid acak.
agglo = AgglomerativeClustering(n_clusters=3)
agglo_labels = agglo.fit_predict(X_scaled)

agg_score = silhouette_score(X_scaled, agglo_labels)
print(f"Silhouette Score Agglomerative: {agg_score:.4f}")

plt.subplot(1, 2, 2)
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=agglo_labels, palette='plasma', s=60)
plt.title('Agglomerative Clustering Result')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(title='Cluster')

plt.tight_layout()
plt.show()
```

### Summary & Analysis
Berdasarkan visualisasi dan Silhouette Score:
1. **K-Means** cenderung memberikan batas yang lebih tegas karena sifatnya yang berbasis partisi melingkar (spherical).
2. **Agglomerative** lebih fleksibel namun sangat bergantung pada kriteria linkage (kedekatan antartitik).
3. Skor yang dihasilkan menunjukkan performa yang cukup mirip, namun K-Means seringkali lebih stabil untuk dataset numerik dengan dimensi rendah seperti ini.
